```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heart-disease.csv")
df.head()
```

```python
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (10,10))
```

```python
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (10,10))

scatter = ax0.scatter(x=df["age"],
                     y=df["chol"],
                     c=df["target"])
ax0.legend(*scatter.legend_elements(), title = "Targets")
ax0.axhline(y=df["chol"].mean(),linestyle ='--')
ax0.set(xlabel="age",ylabel="cholesterol",title="Heart Disease and Cholesterol levels")

sactter = ax1.scatter(x=df["age"],
                     y=df["thalach"],
                     c =df["target"])
ax1.legend(*scatter.legend_elements(),title="Targets")
```
